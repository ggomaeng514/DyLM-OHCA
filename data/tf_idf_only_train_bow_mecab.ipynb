```python
import pandas as pd
import re
import numpy as np

from tqdm import tqdm
```

```python
train_df = pd.read_csv('/Users/dongjoon/Documents/bow/not_stacked/train_json_audio_data_decoder_time_arrest.csv')
sentence = train_df.text.tolist()
```

```python
df = pd.read_csv('/Users/dongjoon/Documents/bow/not_stacked/원본_frequnecy_df_cum.csv',index_col=0)
word_list = df.Word.tolist()
word_list
```

```python
def generate_bows_sen(sentence , dict_list):
    #sentence 
    bag_vectors = np.zeros(len(dict_list))

    for i, word in enumerate(dict_list):
        word_list=[]
        word_list =  re.findall(str(word), str(sentence))
        num_word = len(word_list)
        bag_vectors[i] += num_word
    tmp_bow = bag_vectors.tolist()    
    return tmp_bow # 한 EHR_DX row 에 들어있는 token 개수, bow count한 row -> dict_list 개수 +1 
```

```python
def main(EHR , words_list) :
    bows = []
    for row in tqdm(EHR) :
        bow = generate_bows_sen(row, words_list)
        bows.append(bow)
    
    bow_df = pd.DataFrame(bows)
    bow_df.columns = words_list
    print("export")
    bow_df.to_csv('./원본_bow_df.csv', encoding='utf-8-sig')
```

```python
main(sentence, word_list)
```

```python
import pandas as pd 
import numpy as np
```

```python
train_df = pd.read_csv('/Users/dongjoon/Documents/bow/not_stacked/train_json_audio_data_decoder_time_arrest.csv')
sentence = train_df.text.tolist()
id = train_df[['id']].reset_index(drop=True)
```

```python
tf = pd.read_csv('./원본_bow_df.csv')
tf.drop(['Unnamed: 0'], axis = 1, inplace = True)
```

```python
df = tf.astype(bool).sum(axis = 0)
len(df)
```

```python
import numpy as np

# 문서 개수
D = len(tf)

# Inverse Document Frequency
idf = np.log((D+1) / (df+1)) + 1

tfidf = tf * idf                      
norm = np.linalg.norm(tfidf, axis=1, keepdims=True)
norm[norm == 0] = 1e-10  # Replace zero norms with a small number
tfidf = tfidf / norm
```

```python
idf_df = pd.DataFrame(idf, columns=['idf'])
idf_df.to_csv('원본_train_idf_df.csv',encoding='utf-8-sig')
```

```python
tfidf = pd.concat([tfidf, id], axis=1)
tfidf
```
